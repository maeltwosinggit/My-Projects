```python
import pandas as pd

# List of salutations
salutations = [
    "ASSOC PROF. Ir. Dr.",
    "BRIG. JEN. DATO' IR. HJ.",
    "DATO' Ir. Dr.",
    "DATO' Ir. HJ",
    "DATUK Ir. Hj.",
    "Ir. Dr.",
    "Ir.",
    "PROF. MADYA Ir. Dr.",
    "ASSOC. PROF. IR. DR.",
    "DATO' PROF. Ir. Dr.",
    "DATUK Ir.",
    "DATUK PROF. Ir. Dr.",
    "KAPTEN Ir.",
    "KOL (B) Ir.",
    "KOL. Ir.",
    "KOMANDER Ir.",
    "LT CDR (B)",
    "LT KOL Ir.",
    "MEJ (B) Ir.",
    "MEJAR (B) Ir.",
    "MEJAR Ir.",
    "PROF. Ir. Dr.",
    "PROF. Ir. Dr.",
    "PROF. Ir.",
    "PROF. MADYA Ir.",
    "TAN SRI Ir. Dr.",
    "DATO' Ir."
]


# Read the CSV file into a pandas DataFrame
df = pd.read_csv('bem_mechanical_PE_Practicing_Certificate_List.csv')

# Full names including salutation
full_names = df['Name_name']

# Initialize an empty list to store dictionaries
data = []

# Iterate through each full name
for full_name in full_names:
    # Initialize salutation and name variables
    salutation = ""
    name = full_name
    
    # Check if any salutation appears at the beginning of the full name
    for s in salutations:
        if full_name.startswith(s):
            salutation = s
            # Remove the salutation from the full name
            name = full_name[len(s):].strip()
            break
    
    # Append salutation and name to the list as a dictionary
    data.append({'Full Name': full_name, 'Salutation': salutation, 'Name': name})

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)

df
```

```python
# Define the path for the Excel file
excel_file_path = 'salutation_seperated.xlsx'

# Save the filtered DataFrame to an Excel file
df.to_excel(excel_file_path, index=False)

print("Filtered data saved to", excel_file_path)
```
