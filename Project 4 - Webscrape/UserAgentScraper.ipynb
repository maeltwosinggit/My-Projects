```python
import requests, random
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

```python
response = requests.get('https://explore.whatismybrowser.com/useragents/explore/operating_system_name/windows/')
result = []

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'lxml')

 # Find and extract the first search result link

search_results = soup.find_all('td')
print(search_results)

for search in search_results:
         
    print(f'User Agent = {search.text}')
    result.append([{'User Agent':search.text}])
```

```python
# Flatten the list of dictionaries
flattened_user_agents_data = [d['User Agent'] for sublist in result for d in sublist]

# Create a DataFrame from the flattened data
df = pd.DataFrame(flattened_user_agents_data, columns=['User Agent'])

# Display the DataFrame
df
```

```python
# Define the path for the Excel file
excel_file_path = 'UserAgentScraper.xlsx'

# Save the filtered DataFrame to an Excel file
df.to_excel(excel_file_path, index=False)

print("Result data saved to", excel_file_path)
```
