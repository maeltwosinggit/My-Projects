```python
import pandas as pd
```

```python
# Read the SalesQL Excel files into DataFrames
dfsource1 = pd.read_excel('C:/Users/uadrian/Downloads/MITS/SalesQL Subscription/SalesQL Export Record/SalesQL_contacts_export-2024-05-04-185103.xlsx')
dfsource2 = pd.read_excel('C:/Users/uadrian/Downloads/MITS/SalesQL Subscription/SalesQL Export Record/SalesQL_contacts_export-2024-05-06-172112.xlsx')
dfsource3 = pd.read_excel('C:/Users/uadrian/Downloads/MITS/SalesQL Subscription/SalesQL Export Record/SalesQL_contacts_export-2024-05-09-025526 - Johnny Go.xlsx')

# Concatenate the DataFrames vertically
combined_df = pd.concat([dfsource1, dfsource2, dfsource3], ignore_index=True)
```

```python
# Custom function to extract profile ID from LinkedIn URL
def extract_profile_id(url):
    url = url.split('?')[0]
    url = url.split('#')[0]
    parts = url.split('/')
    return parts[4]

# Apply the custom function to the whole column
combined_df['Profile_ID'] = combined_df['person_linkedin_url'].apply(extract_profile_id)
```

```python
# Read the Excel file into a DataFrame
dftemplate = pd.read_excel('C:/Users/uadrian/Downloads/MITS/Data Research Template.xlsx', sheet_name='Template', header=1)
dftemplate['LinkedIn Profile'] = dftemplate['LinkedIn Profile'].str.replace('%C2%AE', '®')
dftemplate.dropna(subset=['LinkedIn Profile'], inplace=True)

# Apply the custom function to the whole column
dftemplate['Profile_ID'] = dftemplate['LinkedIn Profile'].apply(extract_profile_id)
```

```python
#joining df
join_df = pd.merge(combined_df, dftemplate[['Profile_ID', 'Raw Full Name']], on='Profile_ID', how='left')
join_df[join_df['Raw Full Name'].isna()]
```

```python
df1 = join_df[['Raw Full Name','email1','email1_type', 'email1_status', 'phone1', 'phone1_type','phone1_status']]
df2 = join_df[['Raw Full Name','email2','email2_type', 'email2_status', 'phone2', 'phone2_type','phone2_status']]
df3 = join_df[['Raw Full Name','email3','email3_type', 'email3_status', 'phone3', 'phone3_type','phone3_status']]
df4 = join_df[['Raw Full Name','email4','email4_type', 'email4_status', 'phone4', 'phone4_type','phone4_status']]

# Renaming columns
df1.rename(columns={'email1': 'email', 'email1_type': 'email_type', 'email1_status': 'email_status',
                    'phone1': 'phone', 'phone1_type': 'phone_type', 'phone1_status': 'phone_status'}, inplace=True)

df2.rename(columns={'email2': 'email', 'email2_type': 'email_type', 'email2_status': 'email_status',
                    'phone2': 'phone', 'phone2_type': 'phone_type', 'phone2_status': 'phone_status'}, inplace=True)

df3.rename(columns={'email3': 'email', 'email3_type': 'email_type', 'email3_status': 'email_status',
                    'phone3': 'phone', 'phone3_type': 'phone_type', 'phone3_status': 'phone_status'}, inplace=True)

df4.rename(columns={'email4': 'email', 'email4_type': 'email_type', 'email4_status': 'email_status',
                    'phone4': 'phone', 'phone4_type': 'phone_type', 'phone4_status': 'phone_status'}, inplace=True)

# Combining DataFrames
rearranged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
# Drop rows where both 'email' and 'phone' columns are null
filtered_df = rearranged_df.dropna(subset=['email', 'phone'], how='all')
filtered_df
```

Xlookup process into Python Process

```python
# extracting data from template
dftemplate_raw = pd.read_excel('C:/Users/uadrian/Downloads/MITS/Data Research Template.xlsx', sheet_name='Template', header=1)

# drop last three useless column
# dftemplate_raw = dftemplate_raw.iloc[:, :-3]
dftemplate_raw
```

```python
#Task 1 - From Template Xlookup Contact Email (email type 'personal') with email status 'valid'

# Filter filtered_df for rows where 'email_status' is 'valid' and 'email_type' is 'personal'
filtered_df_t1 = filtered_df[(filtered_df['email_status'] == 'valid') & (filtered_df['email_type'] == 'personal')]
print(f'count rows for filtered_df: {filtered_df.shape}\ncount rows for filtered_df_t1: {filtered_df_t1.shape}')
# Create a copy of dftemplate_raw
dftemplate_t1 = dftemplate_raw.copy()
# Create a dictionary mapping 'Raw Full Name' to 'email' in filtered_df
email_mapping = dict(zip(filtered_df_t1['Raw Full Name'], filtered_df_t1['email']))
# Fill NaN values in 'email' column of dftemplate_t1 using email_mapping
dftemplate_t1['CONTACT EMAIL'] = dftemplate_t1['CONTACT EMAIL'].fillna(dftemplate_t1['Raw Full Name'].map(email_mapping))

#Task 2 - From Template Xlookup Work Email (email type 'main job' & 'job') with email status 'valid'
# Filter filtered_df for rows where 'email_status' is 'valid' and 'email_type' is 'main job' or 'job'
filtered_df_t2 = filtered_df[(filtered_df['email_status'] == 'valid') & (filtered_df['email_type'].isin(['main job', 'job']))]
# Create a copy of dftemplate_raw
dftemplate_t2 = dftemplate_t1.copy()
# Create a dictionary mapping 'Raw Full Name' to 'email' in filtered_df_t2, only for 'main job' and 'job' emails
email_mapping_t2 = dict(zip(filtered_df_t2['Raw Full Name'], filtered_df_t2['email']))
# Fill NaN values in 'WORK EMAIL' column of dftemplate_t2 using email_mapping_t2
dftemplate_t2['WORK EMAIL'] = dftemplate_t2['WORK EMAIL'].fillna(dftemplate_t2['Raw Full Name'].map(email_mapping_t2))

#Task 3 - From Template Xlookup Contact Tel (phone type '') with phone status status 'valid'
# Filter filtered_df for rows where 'phone_status' is 'valid' and 'phone_type' is ''
filtered_df_t3 = filtered_df[(filtered_df['phone_status'] == 'valid')]
# Create a copy of dftemplate_raw
dftemplate_t3 = dftemplate_t2.copy()
# Create a dictionary mapping 'Raw Full Name' to 'phone' in filtered_df_t3, only for phone_type ''
phone_mapping_t3 = dict(zip(filtered_df_t3['Raw Full Name'], filtered_df_t3['phone']))
# Fill NaN values in 'CONTACT TEL' column of dftemplate_t3 using phone_mapping_t3
dftemplate_t3['CONTACT TEL'] = dftemplate_t3['CONTACT TEL'].fillna(dftemplate_t3['Raw Full Name'].map(phone_mapping_t3))

#Task 4 - From Template column 'Raw Full Name', xlookup join_df 'Raw Full Name' and return column 'company_name'
# Create a copy of dftemplate_raw
dftemplate_t4 = dftemplate_t3.copy()
# Create a dictionary mapping 'Raw Full Name' to 'company_name' in join_df
company_mapping = dict(zip(join_df['Raw Full Name'], join_df['company_name']))
# Fill NaN values in 'COMPANY NAME' column of dftemplate_t4 using company_mapping
dftemplate_t4['COMPANY NAME'] = dftemplate_t4['COMPANY NAME'].fillna(dftemplate_t4['Raw Full Name'].map(company_mapping))

#Task 5 - From Template column 'Raw Full Name', xlookup join_df 'Raw Full Name' and return column 'company_website'
# Create a copy of dftemplate_raw
dftemplate_t5 = dftemplate_t4.copy()
# Create a dictionary mapping 'Raw Full Name' to 'company_website' in join_df
website_mapping = dict(zip(join_df['Raw Full Name'], join_df['company_website']))
# Fill NaN values in 'COMPANY WEBSITE' column of dftemplate_t5 using website_mapping
dftemplate_t5['COMPANY WEBSITE'] = dftemplate_t5['COMPANY WEBSITE'].fillna(dftemplate_t5['Raw Full Name'].map(website_mapping))

# dftemplate_t5
```
