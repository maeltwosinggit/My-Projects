```python
import requests, random
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
```

```python
url = "https://www.greenre.org/greere_accredited_professionals"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    print(soup.title.text)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
search_results = soup.find('tbody')

if search_results:
    rows = search_results.find_all('tr')
    extracted_data = []

    for row in rows:
        cells = row.find_all('td')
        data = {
            "No.": cells[0].text.strip(),
            "Name": cells[1].text.strip(),
            "Company": cells[2].text.strip(),
            "GREENREAP No.": cells[3].text.strip(),
            "Email Address": cells[4].text.strip(),
            "Expiry Date": cells[5].text.strip()
        }
        extracted_data.append(data)

    # Convert the extracted data to a DataFrame
    df = pd.DataFrame(extracted_data)
    
    
    # Optionally, save the DataFrame to a CSV file
    # df.to_csv('accredited_professionals.csv', index=False)
else:
    print("No table body found.")
```
