## Extracting from SalesQL to pandas dataframe

reading from excel file SalesQL Raw

```python
# Read the Excel file into a DataFrame
dfSalesQL = pd.read_excel('C:/Users/uadrian/Downloads/MITS/Data Research Template.xlsx', sheet_name='SalesQL Record')
```

## rearrange SalesQL data

```python
df1 = dfSalesQL[['full_name','company_name_cleaned','email1','email1_type', 'email1_status', 'phone1', 'phone1_type','phone1_status']]
df2 = dfSalesQL[['full_name','company_name_cleaned','email2','email2_type', 'email2_status', 'phone2', 'phone2_type','phone2_status']]
df3 = dfSalesQL[['full_name','company_name_cleaned','email3','email3_type', 'email3_status', 'phone3', 'phone3_type','phone3_status']]
df4 = dfSalesQL[['full_name','company_name_cleaned','email4','email4_type', 'email4_status', 'phone4', 'phone4_type','phone4_status']]

# Renaming columns
df1.rename(columns={'email1': 'email', 'email1_type': 'email_type', 'email1_status': 'email_status',
                    'phone1': 'phone', 'phone1_type': 'phone_type', 'phone1_status': 'phone_status'}, inplace=True)

df2.rename(columns={'email2': 'email', 'email2_type': 'email_type', 'email2_status': 'email_status',
                    'phone2': 'phone', 'phone2_type': 'phone_type', 'phone2_status': 'phone_status'}, inplace=True)

df3.rename(columns={'email3': 'email', 'email3_type': 'email_type', 'email3_status': 'email_status',
                    'phone3': 'phone', 'phone3_type': 'phone_type', 'phone3_status': 'phone_status'}, inplace=True)

df4.rename(columns={'email4': 'email', 'email4_type': 'email_type', 'email4_status': 'email_status',
                    'phone4': 'phone', 'phone4_type': 'phone_type', 'phone4_status': 'phone_status'}, inplace=True)

# Combining DataFrames
rearranged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

## Saving the rearranged SalesQL data into Excel

```python
# Define the path for the Excel file
excel_file_path = 'Linkedin_SalesQL1.xlsx'
# Save the filtered DataFrame to an Excel file
rearranged_df.to_excel(excel_file_path, index=False)
print("Filtered data saved to", excel_file_path)
```

## Extracting info from Template into pandas dataframe

```python
# Read the Excel file into a DataFrame
# dftemplate = pd.read_excel('C:/Users/uadrian/Downloads/MITS/Data Research Template.xlsx', sheet_name='Template', header=1)

dftemplate = pd.DataFrame(columns=['NO', 'COMPANY NAME', 'COMPANY ADDRESS', 'COMPANY ACTIVITIES',
       'COMPANY TEL', 'COMPANY EMAIL', 'COMPANY WEBSITE', 'CONTACT SALUTATION',
       'CONTACT NAME', 'CONTACT TEL', 'WORK EMAIL', 'CONTACT EMAIL',
       'POSITION', 'SUBJECT OF INTEREST', 'OTHER NOTES', 'Search Status',
       'Link Reference', 'LinkedIn Profile', 'LinkedIn Connection Status',
       'Google Link Redirect', 'Raw Full Name', 'Added in SalesQL', 'Checking',
       'Unnamed: 23'])
```

Rearranging emails properly

```python
# Drop rows where both 'email' and 'phone' columns are null
filtered_df = rearranged_df.dropna(subset=['email', 'phone'], how='all')
filtered_df = filtered_df.sort_values(by=['company_name_cleaned', 'full_name'])

pretemplate = pd.DataFrame(columns=[
    'NO', 'COMPANY NAME', 'COMPANY ADDRESS', 'COMPANY ACTIVITIES', 'COMPANY TEL', 
    'COMPANY EMAIL', 'COMPANY WEBSITE', 'CONTACT SALUTATION', 'CONTACT NAME', 
    'CONTACT TEL', 'WORK EMAIL', 'CONTACT EMAIL', 'POSITION', 'SUBJECT OF INTEREST', 
    'OTHER NOTES', 'Search Status', 'Link Reference', 'LinkedIn Profile', 
    'LinkedIn Connection Status', 'Google Link Redirect', 'Raw Full Name', 
    'Added in SalesQL', 'Checking', 'Unnamed: 23'
])

# Populate the CONTACT EMAIL and WORK EMAIL columns
pretemplate['CONTACT EMAIL'] = filtered_df.apply(lambda x: x['email'] if x['email_type'] == 'personal' else '', axis=1)
pretemplate['WORK EMAIL'] = filtered_df.apply(lambda x: x['email'] if x['email_type'] in ['main job', 'job'] else '', axis=1)

name_mapping = dict(zip(filtered_df['email'], filtered_df['full_name']))
# Fill NaN values in 'COMPANY NAME' column of dftemplate_t4 using company_mapping
pretemplate['CONTACT NAME'] = pretemplate['CONTACT NAME'].fillna(pretemplate['CONTACT EMAIL'].map(name_mapping))
pretemplate['CONTACT NAME'] = pretemplate['CONTACT NAME'].fillna(pretemplate['WORK EMAIL'].map(name_mapping))

# Drop rows where both 'WORK EMAIL' and 'CONTACT EMAIL' are empty strings
pretemplate = pretemplate[~((pretemplate['WORK EMAIL'] == '') & (pretemplate['CONTACT EMAIL'] == ''))]
pretemplate.replace('', np.nan, inplace=True)
pretemplate[['CONTACT NAME','WORK EMAIL','CONTACT EMAIL']]

# Group by 'CONTACT NAME' and aggregate the emails
stacked_emails = pretemplate.groupby('CONTACT NAME').agg(lambda x: ', '.join(x.dropna()))

# Reset the index if needed
stacked_emails.reset_index(inplace=True)
# Split emails into separate rows
stacked_emails = stacked_emails.assign(
    WORK_EMAIL=stacked_emails['WORK EMAIL'].str.split(', '),
    CONTACT_EMAIL=stacked_emails['CONTACT EMAIL'].str.split(', ')
).explode('WORK EMAIL').explode('CONTACT EMAIL')

# Split comma-separated emails into separate rows
stacked_emails['CONTACT EMAIL'] = stacked_emails['CONTACT EMAIL'].str.split(', ')

# Explode 'CONTACT EMAIL' column
stacked_emails = stacked_emails.explode('CONTACT EMAIL')

# Split comma-separated emails into separate rows
stacked_emails['WORK EMAIL'] = stacked_emails['WORK EMAIL'].str.split(', ')

# Explode 'WORK EMAIL' column
stacked_emails = stacked_emails.explode('WORK EMAIL')

stacked_emails.replace('', np.nan, inplace=True)
stacked_emails = stacked_emails[[
    'NO', 'COMPANY NAME', 'COMPANY ADDRESS', 'COMPANY ACTIVITIES', 'COMPANY TEL', 
    'COMPANY EMAIL', 'COMPANY WEBSITE', 'CONTACT SALUTATION', 'CONTACT NAME', 
    'CONTACT TEL', 'WORK EMAIL', 'CONTACT EMAIL', 'POSITION', 'SUBJECT OF INTEREST', 
    'OTHER NOTES', 'Search Status', 'Link Reference', 'LinkedIn Profile', 
    'LinkedIn Connection Status', 'Google Link Redirect', 'Raw Full Name', 
    'Added in SalesQL', 'Checking', 'Unnamed: 23']]
stacked_emails
```

adding Names that dont have contact infos. (atleast we have their linkedin profiles)

```python
# Select rows where both 'email' and 'phone' columns are NaN
filteredNaN_df = rearranged_df[rearranged_df[['email', 'phone']].isna().all(axis=1)]
filteredNaN_df = filteredNaN_df.drop_duplicates()

# Get the full names from filtered_df
full_names_filtered_df = filtered_df['full_name']

# Check if full names in filteredNaN_df exist in filtered_df
mask = filteredNaN_df['full_name'].isin(full_names_filtered_df)

# Drop rows in filteredNaN_df where full names match those in filtered_df
filteredNaN_df = filteredNaN_df[~mask]

stacked_emails_full = stacked_emails.copy()
# Update 'stacked_emails' with 'full_name' from 'filteredNaN_df' into 'CONTACT NAME' column
stacked_emails_full = stacked_emails_full.merge(filteredNaN_df[['full_name']].rename(columns={'full_name': 'CONTACT NAME'}), left_on='CONTACT NAME', right_on='CONTACT NAME', how='outer')
stacked_emails_full
```
