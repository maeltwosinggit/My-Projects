```python
import glob
import pandas as pd
from dotenv import load_dotenv
import os
from urllib.parse import urlparse

# Load variables from the .env file into the environment
load_dotenv()

# Accessing variables
folder_path = os.getenv('FOLDER_PATH')
```

```python
# Step 3: Use glob to Find CSV Files
csv_files = glob.glob(folder_path + '/*.csv')

# Print out the files found by glob.glob() for debugging
print("CSV Files found:")
print(csv_files)

# Step 4: Load CSV Files Using pandas
dfs = []

for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

# Step 5: Process or Analyze the DataFrames (Optional)
# Example: Concatenate all DataFrames into a single DataFrame
if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    print("Combined DataFrame:")
    print(combined_df.head())
else:
    print("No CSV files found in the specified folder.")
```

Sorting out Links

```python
#Part 1
#select only relevant columns
combined_df_dropduplicate = combined_df.drop_duplicates()
Links = combined_df_dropduplicate[['Link1','Link2','Link3','Link4','Link5','Link6','Link7','Link8']]

# Melt the DataFrame to combine all link columns into one column
melted_links = Links.melt(value_name='Link').drop('variable', axis=1)

# Drop rows with NaN values
combined_links = melted_links.dropna()

# Reset the index
combined_links.reset_index(drop=True, inplace=True)

# The resulting DataFrame with a single column of links
combined_links

#Part 2
# Function to extract the domain from a URL
def extract_domain(url):
    parsed_url = urlparse(url)
    return parsed_url.netloc

# Apply the function to extract domains
combined_links['Domain'] = combined_links['Link'].apply(extract_domain)

# The resulting DataFrame with domains
combined_links

# Display Unique Domains
# combined_links['Domain'].unique()
```

```python
# Extract column names
column_names = combined_df.columns.tolist()

# Create a new DataFrame with the column names
df_columns = pd.DataFrame(column_names, columns=['column'])

# Sort the DataFrame by the column names in ascending order
df_columns_sorted = df_columns.sort_values(by='column').reset_index(drop=True)

# Display the sorted DataFrame
df_columns_sorted
```
