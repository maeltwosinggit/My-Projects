```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import re
import time
from bs4 import BeautifulSoup
```

```python
# Path to your Brave browser
brave_path = r"C:\Program Files\BraveSoftware\Brave-Browser\Application\brave.exe"  # Update this with your actual Brave path

# Path to your ChromeDriver (chromedriver)
chrome_driver_path = r"chromedriver-win64\chromedriver.exe"  # Update this with your actual ChromeDriver path

# URL to scrape
url = 'https://www.propertyguru.com.my/property-listing/arte-mont-kiara-kuala-lumpur-for-sale-by-stanley-ho-41531821'

# Initialize WebDriver
options = Options()
options.binary_location = brave_path
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=options)

# Open the URL
driver.get(url)
time.sleep(5)  # Give the page time to load

# Extract agent name, agency, and contact info
try:
    agent_name = driver.find_element(By.CLASS_NAME, 'agent-name').text
    agency = driver.find_element(By.CLASS_NAME, 'agency').text

    # Find the element with the class name and get the link
    element = driver.find_element(By.CLASS_NAME, 'contact-agent-actions-group')
    link = element.find_element(By.TAG_NAME, 'a')  # Get the first <a> tag inside the element
    whatsapp_link = link.get_attribute('href')  # Extract the href attribute
    contact_number = re.search(r'wa.me/(\d+)', whatsapp_link).group(1)  # Extract only the number part

    # Print the results
    print(f"Agent Name: {agent_name}")
    print(f"Agency: {agency}")
    print(f"Contact Number: {contact_number}")

    # Store the results in a DataFrame
    df = pd.DataFrame([[agent_name, agency, contact_number]], columns=["Agent Name", "Agency", "Contact Number"])

except Exception as e:
    print(f"Error occurred while scraping {url}: {e}")

finally:
    # Quit the driver
    driver.quit()

# # Save the DataFrame to an Excel file
# df.to_excel("agent_info.xlsx", index=False)

# print("Agent information saved to 'agent_info.xlsx'")
```
